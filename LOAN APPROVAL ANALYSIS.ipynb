# Dealing With Null Values

```python
# Check for missing values of dataset
df.isnull().sum()
```

```python
# Filling The Null Values of Gender Column
df.Gender=df.Gender.fillna(df.Gender.mode()[0])
```

```python
# Filling The Null Values of Dependents Column
df.Dependents=df.Dependents.fillna(df.Dependents.mode()[0])
```

```python
# Filling The Null Values of Self_Enployed Column
df.Self_Employed=df.Self_Employed.fillna(df.Self_Employed.mode()[0])
```

```python
# Filling The Null Values of LoanAmount Column
df.LoanAmount=df.LoanAmount.fillna(df.LoanAmount.median())
```

```python
# Filling The Null Values of Credit_History Column
df.Credit_History=df.Credit_History.fillna(df.Credit_History.mode()[0])
```

```python
df.Loan_Amount_Term=df.Loan_Amount_Term.fillna(df.Loan_Amount_Term.median())
```

```python
# Check Null Value After Filling
df.isnull().sum()
```

# Dealing With Outlier In the Dataset

```python
# Checking the Outlier in the ApplicantIncome column of dataset.
plt.figure(figsize=(4,3))
sns.boxplot(df.ApplicantIncome,orient='h')
plt.show()
```

```python
# Removing Outlier from the ApplicantIncome column of the dataset.
Q1,Q3 = df.ApplicantIncome.quantile([0.25,0.75])
```

```python
Q1,Q3
```

```python
#  Finding IQR

IQR = Q3-Q1
```

```python
IQR
```

```python
# Finding Upper Limit and Lower Limit.

UL = Q3+1.5*(IQR)
LL = Q1-1.5*(IQR)
```

```python
UL,LL
```

```python
# TREATING OUTLIERS BY REPLACING THEM WITH
df.ApplicantIncome = np.where(df.ApplicantIncome>UL,UL,df.ApplicantIncome)
```
